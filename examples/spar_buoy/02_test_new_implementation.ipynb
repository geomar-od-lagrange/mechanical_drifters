```python
t_end = 600.0
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from mechanical_drifters.models.spar_buoy_simple import (
    SparBuoyDrifter as SparBuoyDrifterA,
    SparBuoyPhysics as SparBuoyPhysicsA,
)

from mechanical_drifters.models.spar_buoy_simple_reference import (
    SparBuoyDrifter as SparBuoyDrifterB,
    SparBuoyPhysics as SparBuoyPhysicsB,
)
```

```python
drifterA = SparBuoyDrifterA()
print("DrifterA:")
print(drifterA.physics)
```

```python
drifterB = SparBuoyDrifterB()
print("DrifterB:")
print(drifterB.physics)
```

```python
physics_fast_A = SparBuoyPhysicsA(m=1.0, m_tilde=1.0)
drifterA = SparBuoyDrifterA(physics_fast_A)
```

```python
physics_fast_B = SparBuoyPhysicsB(m=1.0, m_tilde=1.0)
drifterB = SparBuoyDrifterA(physics_fast_B)
```

```python
U_surface = 0.3
V_surface = 0.0

def sample_uv_uniform(z):
    z_arr = np.atleast_1d(np.asarray(z, dtype=float))
    N = len(z_arr)
    return np.full(N, U_surface), np.full(N, V_surface)
```

```python
t_out_A, Y_out_A, max_accel_A = drifterA.integrate(
    sample_uv_uniform,
    t_span=(0.0, t_end),
)
drift_A_vel = drifterA.drift_velocity(Y_out_A[-1])

xd_ss_A = float(drift_A_vel[0, 0])
yd_ss_A = float(drift_A_vel[0, 1])

print(f"Surface current:       ({U_surface:.3f}, {V_surface:.3f}) m/s")
print(f"Steady-state drift:    ({xd_ss_A:.6f}, {yd_ss_A:.6f}) m/s")
print(f"Max residual accel:    {max_accel_A:.2e} m/s²")
```

```python
t_out_B, Y_out_B, max_accel_B = drifterB.integrate(
    sample_uv_uniform,
    t_span=(0.0, t_end),
)
drift_B_vel = drifterB.drift_velocity(Y_out_B[-1])

xd_ss_B = float(drift_B_vel[0, 0])
yd_ss_B = float(drift_B_vel[0, 1])

print(f"Surface current:       ({U_surface:.3f}, {V_surface:.3f}) m/s")
print(f"Steady-state drift:    ({xd_ss_B:.6f}, {yd_ss_B:.6f}) m/s")
print(f"Max residual accel:    {max_accel_B:.2e} m/s²")
```

```python
np.testing.assert_almost_equal(Y_out_A[:,:,0], Y_out_B[:,:,0])
np.testing.assert_almost_equal(Y_out_A[:,:,1], Y_out_B[:,:,1])
```
